#### Rhizosphere soil data set from 3 crops.

Reads were subsampled dynamically, kmer trimmed (k 21) using trim-low-abund.py and signatures calculated using sourmash compute -k 21,31,51 and scaled values of  100, 500, 1000, 10000, and 20000

```python
%%bash
for i in 21 31 51
do
~/dev/sourmash/sourmash compare -o rhizosphere_scaled1000_k${i} *1000.sig -k${i}
done
```

```python
D, labels = fig.load_matrix_and_labels('rhizosphere_scaled1000_k21')
D = (D + D.T) / 0.1
_ = fig.plot_composite_matrix(D, labels, vmin=0.0)
numpy.min(D), numpy.max(D)

D, labels = fig.load_matrix_and_labels('rhizosphere_scaled1000_k31')
D = (D + D.T) / 0.1
_ = fig.plot_composite_matrix(D, labels, vmin=0.0)
numpy.min(D), numpy.max(D)

D, labels = fig.load_matrix_and_labels('rhizosphere_scaled1000_k51')
D = (D + D.T) / 0.1
_ = fig.plot_composite_matrix(D, labels, vmin=0.0)
numpy.min(D), numpy.max(D)
```
