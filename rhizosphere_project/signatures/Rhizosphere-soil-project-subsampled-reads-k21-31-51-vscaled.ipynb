#### Rhizosphere soil data set from 3 crops.

Reads were subsampled dynamically, kmer trimmed (k 21) using trim-low-abund.py and signatures calculated using sourmash compute -k 21,31,51 and scaled values of  100, 500, 1000, 10000, and 20000

```python
!~/dev/sourmash/sourmash compare -o rhizosphere_scaled500_k21 *500.sig -k21
!~/dev/sourmash/sourmash compare -o rhizosphere_scaled10000_k51 *10000.sig -k21
!~/dev/sourmash/sourmash compare -o rhizosphere_scaled10000_k51 *10000.sig -k51
```

```python
D, labels = fig.load_matrix_and_labels('rhizosphere_scaled500_k21')
D = (D + D.T) / 0.1
_ = fig.plot_composite_matrix(D, labels, vmin=0.0)
numpy.min(D), numpy.max(D)
```
